```python
df_12 = pd.read_excel('data_last.xlsx', sheet_name='UZ_12')
df_12['well_id'] = 'UZ_12'
df_13 = pd.read_excel('data_last.xlsx', sheet_name='UZ_13')
df_13['well_id'] = 'UZ_13'
df_21 = pd.read_excel('data_last.xlsx', sheet_name='UZ_21')
df_21['well_id'] = 'UZ_21'
df_26 = pd.read_excel('data_last.xlsx', sheet_name='UZ_26V2')
df_26['well_id'] = 'UZ_26'
df_31 = pd.read_excel('data_last.xlsx', sheet_name='UZ_31')
df_31['well_id'] = 'UZ_31'
df_50 = pd.read_excel('data_last.xlsx', sheet_name='UZ_50')
df_50['well_id'] = 'UZ_50'

dfs = [df_12, df_13, df_21, df_26, df_31, df_50]
```

```python
cols = ['well_id', 'TVD', 'GR', 'WNKT', 'DTP', 'DS_diff']
```

```python
# drop unused columns from df
for df in dfs:
    df_cols = list(df.columns)
    df.drop(columns=list(set(df_cols) - set(cols)), inplace=True)
```

```python
df_full = pd.concat(dfs)
df_full.to_pickle('data.pkl')
df_full.set_index(['well_id', 'TVD'])
```
